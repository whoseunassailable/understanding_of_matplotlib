#### Arrange the graph with some particular unit distance

```python
x = [0, 1, 2, 3, 4]
y = [0, 2, 4, 6, 8]
title = 'This is a graph'
plt.rcParams['font.family'] = 'Comic Sans MS'
fontdict = {'fontsize': 28, 'fontweight': 10, 'color': 'GREEN'}
plt.plot(x,y)
plt.title(title, fontdict)
plt.xlabel('Digits define by a single difference!')
plt.ylabel('Digits define by a double difference!')
plt.xticks([0, 1, 2, 3, 4])
plt.yticks([2, 4, 6, 8, 10])
plt.show()
```

#### Legends is a label to the line/lines in a graph

```python
x = [0, 1, 2, 3, 4]
y = [0, 2, 4, 6, 8]
title = 'This is a graph'
plt.rcParams['font.family'] = 'Comic Sans MS'
fontdict = {'fontsize': 28, 'fontweight': 10, 'color': 'GREEN'}
label = 'Blue line'
plt.plot(x,y, label = label)
plt.title(title, fontdict)
plt.xlabel('Digits define by a single difference!')
plt.ylabel('Digits define by a double difference!')
plt.xticks([0, 1, 2, 3, 4])
plt.yticks([2, 4, 6, 8, 10])
plt.legend()
plt.show()
```

#### Increase the width of the line, add a marker and a markerwidth

```python
x = [0, 1, 2, 3, 4]
y = [0, 2, 4, 6, 8]
title = 'This is a graph'
plt.rcParams['font.family'] = 'Comic Sans MS'
fontdict = {'fontsize': 28, 'fontweight': 10, 'color': 'GREEN'}
label = 'Blue line'
linewidth = 2
color = 'RED'
marker = '.'
markersize = 20
markeredgecolor = 'WHITE'
linestyle = '--'
plt.plot(x,y, label = label, linewidth = linewidth, marker = marker, color = color, markersize = markersize, markeredgecolor = markeredgecolor, linestyle = linestyle)
plt.title(title, fontdict)
plt.xlabel('Digits define by a single difference!')
plt.ylabel('Digits define by a double difference!')
plt.xticks([0, 1, 2, 3, 4])
plt.yticks([2, 4, 6, 8, 10])
plt.legend()
plt.show()
```

#### Shorthand notation for color marker and markerwidth
#### fmt = '[color][marker][line]'

```python
x = [0, 1, 2, 3, 4]
y = [0, 2, 4, 6, 8]

title = 'This is a graph'
plt.rcParams['font.family'] = 'Comic Sans MS'
fontdict = {'fontsize': 28, 'fontweight': 10, 'color': 'GREEN'}
label = 'Blue line'
linewidth = 2
markersize = 20
markeredgecolor = 'BLACK'
plt.plot(x,y, 'r.--', label = label, linewidth = linewidth, markersize = markersize, markeredgecolor = markeredgecolor)
plt.title(title, fontdict)
plt.xlabel('Digits defined by a single difference!')
plt.ylabel('Digits defined by a double difference!')
plt.xticks([0, 1, 2, 3, 4])
plt.yticks([2, 4, 6, 8, 10])
plt.legend()
plt.show()
```

#### Adding another line

```python
x = [0, 1, 2, 3, 4]
y = [0, 2, 4, 6, 8]

title = 'This is a graph'
plt.rcParams['font.family'] = 'Comic Sans MS'
fontdict = {'fontsize': 28, 'fontweight': 10, 'color': 'GREEN'}
label = 'Red line'
linewidth = 2
markersize = 20
markeredgecolor = 'BLACK'
plt.plot(x,y, 'r.--', label = label, linewidth = linewidth, markersize = markersize, markeredgecolor = markeredgecolor)
plt.title(title, fontdict)
plt.xlabel('Digits defined by a single difference!')
plt.ylabel('Digits defined by a double difference!')
plt.xticks([0, 1, 2, 3, 4])
plt.yticks([2, 4, 6, 8, 10])

# Line number two 
x2 = np.arange(0, 4.5, 0.5)
y2 = x2**2
label = 'Green line'
linewidth = 2
markersize = 20
markeredgecolor = 'BLACK'
plt.plot(x2[:5], y2[:5], 'g.--', label = label, linewidth = linewidth, markersize = markersize, markeredgecolor = markeredgecolor)
plt.plot(x2[4:], y2[4:], 'g.-', linewidth = linewidth, markersize = markersize, markeredgecolor = markeredgecolor)
plt.title(title, fontdict)
plt.xlabel('Digits defined by a single difference!')
plt.ylabel('Digits defined by a double difference!')
plt.xticks([0, 1, 2, 3, 4])
plt.yticks([2, 4, 6, 8, 10])

plt.legend()
plt.show()
```

#### Adding figsize, changing fontsize accordingly for labels of x and y axis, adding legends and so on

```python
x = [0, 1, 2, 3, 4]
y = [0, 2, 4, 6, 8]

plt.figure(figsize = (3.5, 2), dpi = 300)

title = 'This is a graph'
plt.rcParams['font.family'] = 'Comic Sans MS'
fontsize = 10
font_size_label_axis = 5
fontweight = 5
fontdict = {'fontsize': fontsize, 'fontweight': fontweight, 'color': 'GREEN'}
label = 'Red line'
linewidth = 2
markersize = 10
markeredgecolor = 'BLACK'
plt.plot(x,y, 'r.--', label = label, linewidth = linewidth, markersize = markersize, markeredgecolor = markeredgecolor)
plt.title(title, fontdict)
plt.xlabel('Digits defined by a single difference!', fontsize = font_size_label_axis)
plt.ylabel('Digits defined by a double difference!', fontsize = font_size_label_axis)
plt.xticks([0, 1, 2, 3, 4])
plt.yticks([2, 4, 6, 8, 10])

# Line number two 
x2 = np.arange(0, 4.5, 0.5)
y2 = x2**2
label = 'Green line'
linewidth = 2
markersize = 10
markeredgecolor = 'BLACK'
plt.plot(x2[:5], y2[:5], 'g.--', label = label, linewidth = linewidth, markersize = markersize, markeredgecolor = markeredgecolor)
plt.plot(x2[4:], y2[4:], 'g.-', linewidth = linewidth, markersize = markersize, markeredgecolor = markeredgecolor)
plt.title(title, fontdict)
plt.xticks([0, 1, 2, 3, 4])
plt.yticks([2, 4, 6, 8, 10])
plt.tick_params(axis='x', labelsize=6)  # X-axis tick labels
plt.tick_params(axis='y', labelsize=6)  # Y-axis tick labels
plt.legend(fontsize = font_size_label_axis)
plt.show()
```

#### BAR GRAPH

```python
x = ['A', 'B', 'C']
y = [1, 4, 2]

plt.bar(x, y)
plt.show()

```

#### Design the bars with set_hatch() function

```python
x = ['A', 'B', 'C']
y = [1, 4, 2]

bar_graph = plt.bar(x, y)

patterns = ['/', 'O', '*'] # the patterns variable to define the different patterns 

for i in range(len(bar_graph)):
    bar_graph[i].set_hatch(patterns[i]) # Use the set_hatch() function to set the designs from the patterns variable

plt.show()

```

#### Save the figure with as always 300 dpi for better resolution

```python
x = ['A', 'B', 'C']
y = [1, 4, 2]

bar_graph = plt.bar(x, y)

patterns = ['/', 'O', '*'] # the patterns variable to define the different patterns 

for i in range(len(bar_graph)):
    bar_graph[i].set_hatch(patterns[i]) # Use the set_hatch() function to set the designs from the patterns variable

plt.figure(figsize=(6, 4), dpi=300) # Setting the dpi to 300
plt.savefig(bar_graph.png) # Saving it in a png format
plt.show()

```
