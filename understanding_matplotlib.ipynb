#### Legends is a label to the line/lines in a graph

```python
x = [0, 1, 2, 3, 4]
y = [0, 2, 4, 6, 8]
title = 'This is a graph'
plt.rcParams['font.family'] = 'Comic Sans MS'
fontdict = {'fontsize': 28, 'fontweight': 10, 'color': 'GREEN'}
label = 'Blue line'
plt.plot(x,y, label = label)
plt.title(title, fontdict)
plt.xlabel('Digits define by a single difference!')
plt.ylabel('Digits define by a double difference!')
plt.xticks([0, 1, 2, 3, 4])
plt.yticks([2, 4, 6, 8, 10])
plt.legend()
plt.show()
```

#### Increase the width of the line, add a marker and a markerwidth

```python
x = [0, 1, 2, 3, 4]
y = [0, 2, 4, 6, 8]
title = 'This is a graph'
plt.rcParams['font.family'] = 'Comic Sans MS'
fontdict = {'fontsize': 28, 'fontweight': 10, 'color': 'GREEN'}
label = 'Blue line'
linewidth = 2
color = 'RED'
marker = '.'
markersize = 20
markeredgecolor = 'WHITE'
linestyle = '--'
plt.plot(x,y, label = label, linewidth = linewidth, marker = marker, color = color, markersize = markersize, markeredgecolor = markeredgecolor, linestyle = linestyle)
plt.title(title, fontdict)
plt.xlabel('Digits define by a single difference!')
plt.ylabel('Digits define by a double difference!')
plt.xticks([0, 1, 2, 3, 4])
plt.yticks([2, 4, 6, 8, 10])
plt.legend()
plt.show()
```

#### Adding another line

```python
x = [0, 1, 2, 3, 4]
y = [0, 2, 4, 6, 8]

title = 'This is a graph'
plt.rcParams['font.family'] = 'Comic Sans MS'
fontdict = {'fontsize': 28, 'fontweight': 10, 'color': 'GREEN'}
label = 'Red line'
linewidth = 2
markersize = 20
markeredgecolor = 'BLACK'
plt.plot(x,y, 'r.--', label = label, linewidth = linewidth, markersize = markersize, markeredgecolor = markeredgecolor)
plt.title(title, fontdict)
plt.xlabel('Digits defined by a single difference!')
plt.ylabel('Digits defined by a double difference!')
plt.xticks([0, 1, 2, 3, 4])
plt.yticks([2, 4, 6, 8, 10])

# Line number two 
x2 = np.arange(0, 4.5, 0.5)
y2 = x2**2
label = 'Green line'
linewidth = 2
markersize = 20
markeredgecolor = 'BLACK'
plt.plot(x2[:5], y2[:5], 'g.--', label = label, linewidth = linewidth, markersize = markersize, markeredgecolor = markeredgecolor)
plt.plot(x2[4:], y2[4:], 'g.-', linewidth = linewidth, markersize = markersize, markeredgecolor = markeredgecolor)
plt.title(title, fontdict)
plt.xlabel('Digits defined by a single difference!')
plt.ylabel('Digits defined by a double difference!')
plt.xticks([0, 1, 2, 3, 4])
plt.yticks([2, 4, 6, 8, 10])

plt.legend()
plt.show()
```

#### Adding figsize, changing fontsize accordingly for labels of x and y axis, adding legends and so on

```python
x = [0, 1, 2, 3, 4]
y = [0, 2, 4, 6, 8]

plt.figure(figsize = (3.5, 2), dpi = 300)

title = 'This is a graph'
plt.rcParams['font.family'] = 'Comic Sans MS'
fontsize = 10
font_size_label_axis = 5
fontweight = 5
fontdict = {'fontsize': fontsize, 'fontweight': fontweight, 'color': 'GREEN'}
label = 'Red line'
linewidth = 2
markersize = 10
markeredgecolor = 'BLACK'
plt.plot(x,y, 'r.--', label = label, linewidth = linewidth, markersize = markersize, markeredgecolor = markeredgecolor)
plt.title(title, fontdict)
plt.xlabel('Digits defined by a single difference!', fontsize = font_size_label_axis)
plt.ylabel('Digits defined by a double difference!', fontsize = font_size_label_axis)
plt.xticks([0, 1, 2, 3, 4])
plt.yticks([2, 4, 6, 8, 10])

# Line number two 
x2 = np.arange(0, 4.5, 0.5)
y2 = x2**2
label = 'Green line'
linewidth = 2
markersize = 10
markeredgecolor = 'BLACK'
plt.plot(x2[:5], y2[:5], 'g.--', label = label, linewidth = linewidth, markersize = markersize, markeredgecolor = markeredgecolor)
plt.plot(x2[4:], y2[4:], 'g.-', linewidth = linewidth, markersize = markersize, markeredgecolor = markeredgecolor)
plt.title(title, fontdict)
plt.xticks([0, 1, 2, 3, 4])
plt.yticks([2, 4, 6, 8, 10])
plt.tick_params(axis='x', labelsize=6)  # X-axis tick labels
plt.tick_params(axis='y', labelsize=6)  # Y-axis tick labels
plt.legend(fontsize = font_size_label_axis)
plt.show()
```

#### BAR GRAPH

```python
x = ['A', 'B', 'C']
y = [1, 4, 2]

plt.bar(x, y)
plt.show()

```

#### Design the bars with set_hatch() function

```python
x = ['A', 'B', 'C']
y = [1, 4, 2]

bar_graph = plt.bar(x, y)

patterns = ['/', 'O', '*'] # the patterns variable to define the different patterns 

for i in range(len(bar_graph)):
    bar_graph[i].set_hatch(patterns[i]) # Use the set_hatch() function to set the designs from the patterns variable

plt.show()

```

#### Save the figure with as always 300 dpi for better resolution

```python
x = ['A', 'B', 'C']
y = [1, 4, 2]

bar_graph = plt.bar(x, y)

patterns = ['/', 'O', '*'] # the patterns variable to define the different patterns 

for i in range(len(bar_graph)):
    bar_graph[i].set_hatch(patterns[i]) # Use the set_hatch() function to set the designs from the patterns variable

plt.figure(figsize=(6, 4), dpi=300) # Setting the dpi to 300
plt.savefig(bar_graph.png) # Saving it in a png format
plt.show()

```

### Line Graph

```python
gas = pd.read_csv('gas_prices.csv')
gas
```

#### Displaying a graph

```python
plt.plot(gas.Year, gas.USA)
plt.plot(gas.Year, gas.Canada)

plt.show()
```

#### Adding a title and legend

```python
plt.title("Gas prices over time") # adding the title

plt.plot(gas.Year, gas.USA, label = "United States of America") # Label property is used to display legends
plt.plot(gas.Year, gas.Canada, label = "Canada")
plt.plot(gas.Year, gas['South Korea'], label = "South Korea") # using .South Korea will throw an error so use a array format to access it or maybe change the column name beforehand

plt.legend()
plt.show()
```

#### Adding ticks to the x-axis

```python
plt.title("Gas prices over time") 

plt.plot(gas.Year, gas.USA, label = "United States of America")
plt.plot(gas.Year, gas.Canada, label = "Canada")
plt.plot(gas.Year, gas['South Korea'], label = "South Korea") 

plt.xticks(gas.Year[::3]) #Adding ticks every 4 years

plt.legend()
plt.show()
```

#### Adding the points and markers and line style

```python
plt.title("Gas prices over time") 

markersize = 15
markeredgecolor = 'Black'
plt.plot(gas.Year, gas.USA, 'b.--', label = "United States of America" , markersize = markersize, markeredgecolor = markeredgecolor)
plt.plot(gas.Year, gas.Canada, 'r.--', label = "Canada", markersize = markersize, markeredgecolor = markeredgecolor)
plt.plot(gas.Year, gas['South Korea'], 'g.--', label = "South Korea", markersize = markersize, markeredgecolor = markeredgecolor) 

plt.xticks(gas.Year[::3])

plt.legend()
plt.show()
```

#### Adding x and y labels to the data

```python
plt.title("Gas prices over time") 

markersize = 15
markeredgecolor = 'Black'

plt.plot(gas.Year, gas.USA, 'b.--', label = "United States of America" , markersize = markersize, markeredgecolor = markeredgecolor)
plt.plot(gas.Year, gas.Canada, 'r.--', label = "Canada", markersize = markersize, markeredgecolor = markeredgecolor)
plt.plot(gas.Year, gas['South Korea'], 'g.--', label = "South Korea", markersize = markersize, markeredgecolor = markeredgecolor) 

plt.xticks(gas.Year[::3])
plt.xlabel('Year')
plt.ylabel('Prices in USD')

plt.legend()
plt.show()
```

#### Plotting all the countries using a for loop


```python
plt.title("Gas prices over time") 

markersize = 15
markeredgecolor = 'Black'

# use for loop to get into the country_gas_data 
for country in gas.columns[1:]:
    plt.plot(gas.Year, gas[country], label = country, markersize = markersize, markeredgecolor = markeredgecolor)

plt.xticks(gas.Year[::3].to_list() + [2011])
plt.xlabel('Year')
plt.ylabel('Prices in USD')

plt.legend()
plt.show()
```

#### Adding fontdict

```python
fontdict = {'fontsize': 28, 'fontweight': 10, 'color': 'GREEN'}

plt.title("Gas prices over time", fontdict = fontdict) 

markersize = 15
markeredgecolor = 'Black'

for country in gas.columns[1:]:
    plt.plot(gas.Year, gas[country], label = country, markersize = markersize, markeredgecolor = markeredgecolor)

plt.xticks(gas.Year[::3].to_list() + [2011])
plt.xlabel('Year')
plt.ylabel('Prices in USD')

plt.legend()
plt.show()
```

#### Saving the plot

```python
fontdict = {'fontsize': 28, 'fontweight': 10, 'color': 'GREEN'}

plt.figure(figsize = (5, 8), dpi=300) # Giving the intended dpi of 300 as always
plt.title("Gas prices over time", fontdict = fontdict) 

markersize = 15
markeredgecolor = 'Black'

for country in gas.columns[1:]:
    plt.plot(gas.Year, gas[country], label = country, markersize = markersize, markeredgecolor = markeredgecolor)

plt.xticks(gas.Year[::3].to_list() + [2011])
plt.xlabel('Year')
plt.ylabel('Prices in USD')
plt.savefig("Gas_price_figure", dpi = 300)
plt.legend()
plt.show()
```

### Fifa Data

```python
# Loading the data
fifa = pd.read_csv('fifa_data.csv')
fifa
```

#### Plotting a histogram

```python
bins = np.arange(0,100,5) # Created a bins variable that can range from 0 - 100

plt.hist(fifa.Overall, bins = bins) # passed the bins variable as an argument to the bins property

plt.show()
```

#### Adding x tick

```python
bins = np.arange(0,100,5)

plt.hist(fifa.Overall, bins = bins)

plt.xticks(bins.tolist() + [100]) # Using xtick() to add the range value properly

plt.show()
```

#### Adding x and y label and a title

```python
bins = np.arange(0,100,5)
fontdict = {'fontsize': 18, 'fontweight': 10, 'color': 'GREEN'}

fontdictforlabels = {'fontsize': 14, 'fontweight': 5, 'color': 'Black'}

plt.hist(fifa.Overall, bins = bins)
plt.title("Distribution of player skill in FIFA 2018!", fontdict=fontdict)

plt.xticks(bins.tolist() + [100]) # Using xtick() to add the range value properly

plt.xlabel("Overall ratings!", fontdict=fontdictforlabels)
plt.ylabel("Number of players", fontdict=fontdictforlabels)

plt.show()
```

#### Saving the figure

```python
bins = np.arange(0,100,5)
plt.figure(figsize=(5,8), dpi = 300)
fontdict = {'fontsize': 18, 'fontweight': 10, 'color': 'GREEN'}

fontdictforlabels = {'fontsize': 14, 'fontweight': 5, 'color': 'Black'}

plt.hist(fifa.Overall, bins = bins)
plt.title("Distribution of player skill in FIFA 2018!", fontdict=fontdict)

plt.xticks(bins.tolist() + [100]) # Using xtick() to add the range value properly

plt.xlabel("Overall ratings!", fontdict=fontdictforlabels)
plt.ylabel("Number of players", fontdict=fontdictforlabels)
plt.savefig('histogram_for_fifa_2018', dpi = 300)
plt.show()
```
